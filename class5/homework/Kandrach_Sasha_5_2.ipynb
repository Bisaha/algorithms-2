```python
def weight_estimation(individual_height, individual_gender): 
    Intercept= -244.923503
    Male= 19.377711 
    Slope= 5.976941
    if individual_gender == 'Male':
        return (Intercept + Male + (Slope * float(individual_height)))
    else:
        return (Intercept + (Slope * float(individual_height)))
```
